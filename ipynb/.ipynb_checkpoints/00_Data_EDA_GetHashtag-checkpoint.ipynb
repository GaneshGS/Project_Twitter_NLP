```python
def tweet_cleaner(text):
    text = text.lower()
    text = re.sub("'","''", text)
    text = re.sub("{","\{",text)
    text = re.sub("}","\}",text)
    text = re.sub(r'http\S+', '',text)
    #text = re.sub(r'#\S+', '',text)
    text = re.sub(r'@\S+', '',text)
    text = re.sub('\s+',' ',text)
    text = re.sub('\n',' ',text)

    return text
```

```python
get_hashtags = lambda tweet: " ".join([i for i in tweet.split() if ('#' in i)])
for j in tqdm(range(100)):
    tweet_id = df['id'][j]
    hashtags = get_hashtags(df['tweet'][j])
    #hashtags2 = hashtags.replace('#','')
    #hashtags2 = re.sub('\.+','',hashtags2)
    hashtags2 = re.sub('\W',' ',hashtags)
    hashtags2 = re.sub('\s+',' ',hashtags2)
    
    if len(hashtags2) > 1:
        print(hashtags2)    
```

```python
conn = pg2.connect(host = this_host, user = this_user,password = this_password)
cur = conn.cursor(cursor_factory=RealDictCursor)
get_hashtags = lambda tweet: " ".join([i for i in tweet.split() if ('#' in i)])
for j in tqdm(range(len(df))):
    tweet_id = df['id'][j]
    hashtags = get_hashtags(df['tweet'][j])
    #hashtags2 = hashtags.replace('#','')
    #hashtags2 = re.sub('\.+','',hashtags2)
    hashtags2 = re.sub('\W',' ',hashtags)
    hashtags2 = re.sub('\s+',' ',hashtags2)
    
    if len(hashtags2) > 1:
        sql_update = '''
        update tweets 
        set hashtags = '{}'
        where id = '{}';
        commit;'''.format(hashtags2,tweet_id)
        cur.execute(sql_update)
conn.close()    
```
